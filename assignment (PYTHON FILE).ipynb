EXTRACTION OF TITTLE AND CONTENT

```python
# Function to extract title and content
def extract_title_and_content(url):
    try:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')
        article_title = soup.find('h1', class_='entry-title').text
        article_content = soup.find('div', class_='td-post-content').text.replace("\n", "")
        return article_title, article_content
    except Exception as e:
        return None, "Content not available"  # Default value for content
```

```python
# Create an empty DataFrame for titles and contents
df_results = pd.DataFrame(columns=['URL', 'Title', 'Content'])
```

```python
# Assuming you have a list of URLs named "urls" and a DataFrame named "data" with columns "Title" and "Content"
df_results = pd.DataFrame(columns=['URL', 'Title', 'Content'])
for i, url in enumerate(urls):
    if len(df_results) >= 100:
        break

    title, content = extract_title_and_content(url)
    if title and content:
        # Save data to a text file
        filename = f"{i}.txt"
        with open(filename, 'w', encoding='utf-8') as file:
            file.write(f"Title: {title}\n\n")
            file.write(f"Content:\n{content}\n")
        print(f"Saved data for URL {url} in {filename}")

        # Append data to the results DataFrame
        df_results = pd.concat([df_results, pd.DataFrame([[url, title, content]],
                                                         columns=['URL', 'Title', 'Content'])],
                               ignore_index=True)
    else:
        print(f"Error processing URL {url}")

# Drop rows with missing title or content
df_results.dropna(subset=['Title', 'Content'], inplace=True)

# Print the resulting DataFrame
print(df_results[['URL', 'Title', 'Content']])
```

CALCULATION OF AVG_SENTENCE_LENGTH  PERCENT_COMPLEX_WORDS   FOG_INDEX  COMPLEX_WORD_COUNT

```python
from textstat import flesch_reading_ease
# Sample data (replace this with your actual data)
data2 = [art]

df6 =pd.DataFrame(art,columns=['processed_content'])

def calculate_metrics(text):
    sentences = text.split('.')
    total_sentences = len(sentences)
    total_words = len(text.split())
    avg_sentence_length = total_words / total_sentences if total_sentences != 0 else 0

    complex_words_count = sum(1 for word in text.split() if flesch_reading_ease(word) < 60)
    percentage_complex_words = (complex_words_count / total_words) * 100 if total_words != 0 else 0

    fog_index = 0.4 * (avg_sentence_length + percentage_complex_words)

    return avg_sentence_length, percentage_complex_words, fog_index, complex_words_count

df6[['AVG_SENTENCE_LENGTH', 'PERCENT_COMPLEX_WORDS', 'FOG_INDEX', 'COMPLEX_WORD_COUNT']] \
        = df6.apply(lambda row: pd.Series(calculate_metrics(row['processed_content'])), axis=1)
# Drop the 'processed_content' column
df6.drop(columns=['processed_content'], inplace=True)

print(df6)
```

CALCULATION OF   Word Count  Syllables Per Word  Personal Pronouns  Avg Word Length

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from textstat import syllable_count

# Load your data into a DataFrame (replace 'your_data.csv' with your actual file path)
df7 =pd.DataFrame(art,columns=['processed_content'])

# Define a function to calculate metrics for a given text
def calculate_metrics(text):
    # Tokenize the text into words
    words = word_tokenize(text)

    # Calculate word count
    word_count = len(words)

    # Calculate syllables per word
    total_syllables = sum(syllable_count(word) for word in words)
    syllables_per_word = total_syllables / max(word_count, 1)

    # Count personal pronouns (you can customize this list based on your requirements)
    personal_pronouns = sum(1 for word in words if word.lower() in ['i', 'me', 'my', 'mine', 'we', 'us', 'our', 'ours'])

    # Calculate average word length
    avg_word_length = sum(len(word) for word in words) / max(word_count, 1)

    return pd.Series([word_count, syllables_per_word, personal_pronouns, avg_word_length],
                     index=['Word Count', 'Syllables Per Word', 'Personal Pronouns', 'Avg Word Length'])

# Apply the function to each row of your data
df7[['Word Count', 'Syllables Per Word', 'Personal Pronouns', 'Avg Word Length']] = df7['processed_content'].apply(calculate_metrics)

# Drop the 'processed_content' column
df7.drop(columns=['processed_content'], inplace=True)

# Print the updated DataFrame
print(df7)
```
